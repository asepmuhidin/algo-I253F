## SOAL 1 : Rumus Lingkaran
$$
L = \pi r^2
$$
## Rumus Kecepatan
$$
v = s / t
$$

```python
r=10
s=100
t=20

L=3.14*r**2
v=s/t


print(f"Luas lingkaran dengan jar-jari {r} adalah {L}")
print(f"Kecepatan dengan jarak {s} dan waktu {t} adalah {v}")
```

## SOAL 2:  Buat kode program :
Seorang kasir di toko “NextMart” ingin
menghitung total harga belanja pelanggan.
Output Program:
* 3 Kopi    4500    13.500
* 1 Minyak  45000   45.000
* ========================
* Total Belanja     58.500


```python
item1="Kopi"
harga1=4500
jml_beli1=3

item2="Minyak"
harga2=45000
jml_beli2=1

total_belanja=(harga1*jml_beli1)+(harga2*jml_beli2)

print(f"{jml_beli1} {item1}      Rp.{harga1*jml_beli1:,}")
print(f"{jml_beli2} {item2}      Rp.{harga2*jml_beli2:,}")
print("="*25)
print(f"Total Belanja : Rp.{total_belanja:,}")


```
