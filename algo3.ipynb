```python
a,b,c = 1,4,12

det=b**2-(4*a*c)
#Sifat akar: real(det>0), kembar(det=0), imajiner(det<0)
if det>0:
  sifat_akar="Real berlainan"
  x1=(-b+math.sqrt(b**2-4*a*c))/(2*a)
  x2=(-b-math.sqrt(b**2-4*a*c))/(2*a)
elif det==0:
  sifat_akar="Kembar"
  x1=(-b+math.sqrt(b**2-4*a*c))/(2*a)
  x2=x1
else:
  sifat_akar="Imajiner"

print(f"Sifat akar : {sifat_akar}")

if sifat_akar!="Imajiner":
  print(f"x1={x1}, x2={x2}")

```
